# **Ánalisis**

# **Aplicar DBH**


---



```python
cuadrante1["DBH"] = round(cuadrante1["Diametro Altura Pecho (DAP)"].apply(lambda a: a / np.pi),1)
cuadrante2["DBH"] =round(cuadrante2["Diametro Altura Pecho (DAP)"].apply(lambda a: a / np.pi),1)
cuadrante3["DBH"] = round(cuadrante3["Diametro Altura Pecho (DAP)"].apply(lambda a: a / np.pi),1)
cuadrante4["DBH"] = round(cuadrante4["Diametro Altura Pecho (DAP)"].apply(lambda a: a / np.pi),1)
cuadrante5["DBH"] = round(cuadrante5["Diametro Altura Pecho (DAP)"].apply(lambda a: a / np.pi),1)
```

---
# **Calcular Radio**
---

```python
cuadrante1["Radio"] = round(cuadrante1["DBH"].apply(lambda c: c / 2),1)
cuadrante2["Radio"] =round(cuadrante2["DBH"].apply(lambda c: c / 2),1)
cuadrante3["Radio"] = round(cuadrante3["DBH"].apply(lambda c: c / 2),1)
cuadrante4["Radio"] = round(cuadrante4["DBH"].apply(lambda c: c / 2),1)
cuadrante5["Radio"] = round(cuadrante5["DBH"].apply(lambda c: c / 2),1)
```

---
# **Área Basal**
---


```python
def basal(radio):
    return round(np.pi * (radio ** 2),1)
```

```python
cuadrante1["Área Basal (cm²)"] = cuadrante1["DBH"].apply(basal)
cuadrante2["Área Basal (cm²)"] = cuadrante2["DBH"].apply(basal)
cuadrante3["Área Basal (cm²)"] = cuadrante3["DBH"].apply(basal)
cuadrante4["Área Basal (cm²)"] = cuadrante4["DBH"].apply(basal)
cuadrante5["Área Basal (cm²)"] = cuadrante5["DBH"].apply(basal)
```

---
# **Área Basal en Metros Cuadrados**
---

```python
cuadrante1["Área Basal (m²)"] = round(cuadrante1["Área Basal (cm²)"].apply(lambda z: z / 1000),1)
cuadrante2["Área Basal (m²)"] =round(cuadrante2["Área Basal (cm²)"].apply(lambda z: z / 1000),1)
cuadrante3["Área Basal (m²)"] = round(cuadrante3["Área Basal (cm²)"].apply(lambda z: z / 1000),1)
cuadrante4["Área Basal (m²)"] = round(cuadrante4["Área Basal (cm²)"].apply(lambda z: z / 1000),1)
cuadrante5["Área Basal (m²)"] = round(cuadrante5["Área Basal (cm²)"].apply(lambda z: z / 1000),1)
```

---
# **Área Basal en Hectareas**
---

```python
cuadrante1["Área Basal (m²/ha)"] = round(cuadrante1["Área Basal (m²)"].apply(lambda d: d * 10000),1)
cuadrante2["Área Basal (m²/ha)"] = round(cuadrante2["Área Basal (m²)"].apply(lambda d: d * 10000),1)
cuadrante3["Área Basal (m²/ha)"] = round(cuadrante3["Área Basal (m²)"].apply(lambda d: d * 10000),1)
cuadrante4["Área Basal (m²/ha)"] = round(cuadrante4["Área Basal (m²)"].apply(lambda d: d * 10000),1)
cuadrante5["Área Basal (m²/ha)"] = round(cuadrante5["Área Basal (m²)"].apply(lambda d: d * 10000),1)
```

```python
from pyproj import Proj, transform
data = {
    'X': [484475.3, 484504.6, 484563, 484592.3, 483978.4, 484007.7, 484007.7, 484066.2, 485911.5],
    'Y': [2100495.9, 2100465.1, 2100465.1, 2100465.0, 2100619.2, 2100680.7, 2100711.4, 2100742.1, 2105166.5],
    'Salinidad Registrada': [25, 20, 19, 25, 5, 5, 5, 18, 48]
}

df2023 = pd.DataFrame(data)
def utm_to_latlon(x, y):
    in_proj = Proj(init='epsg:32619')
    out_proj = Proj(init='epsg:4326')
    lon, lat = transform(in_proj, out_proj, x, y)
    return lat, lon

# Aplicar la función y agregar las nuevas columnas al DataFrame
df2023['Latitud'], df2023['Longitud'] = zip(*df2023.apply(lambda row: utm_to_latlon(row['X'], row['Y']), axis=1))
print(df2023)
```
